scatter plot

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([5,7,8,7,3,5,9,11,12,9,6])
y = np.array([99,33,88,111,86,95,76,78,65,99,66])

plt.scatter(x,y,color = 'red')
plt.show
```

```python
x = np.array([5,7,8,7,3,5,9,11,12,9,6])
y = np.array([99,33,88,111,86,95,76,78,65,99,66])

plt.scatter(x,y,color = 'red')

x = np.array([2,2,8,1,15,8,12,9,7,6,9])
y = np.array([100,105,111,115,90,80,77,54,112,91,80])

plt.scatter(x,y)

plt.show()
```

BAR PLOTING

```python
x = np.array(['A','B','C','D'])
y = np.array([1,7,5,8])

plt.bar(x,y)
plt.show()
```

HORIZONTAL BARS

```python
x = np.array(['A','B','C','D'])
y = np.array([1,7,5,8])

plt.barh(x,y)
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([1,7,5,8])

plt.bar(x,y,color = 'green')
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([1,7,5,8])


plt.bar(x,y,color = 'green')
plt.xlabel("name")
plt.ylabel("age",rotation = 360)
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([1,7,5,8])


plt.bar(x,y,color = 'green',width = 0.5)
plt.xlabel("name")
plt.ylabel("age",rotation = 360)
plt.show()
```

HISTOGRAM

```python
x = np.random.normal(170 ,18,250)

plt.hist(x)
plt.show()
```

CREATING PIE CHART

```python
x = np.array([35,25,23,15])

plt.pie(x)
plt.show()
```

```python
x = np.array([35,25,23,15])

labels = ['apple','orange','banana','cherries']

plt.pie(x ,labels = labels)

plt.show()
```

```python
x = np.array([35,25,23,15])

labels = ['apple','orange','banana','cherries']
explode =[0,0,0.4,0]

plt.pie(x ,labels = labels ,explode = explode)

plt.show()
```

```python
x = np.array([35,25,23,15])

labels = ['apple','orange','banana','cherries']
c = ['red','orange','yellow' ,'pink']

plt.pie(x ,labels = labels,colors = c)
plt.legend()

plt.show()
```

```python
plt.figure(figsize =(10,10))
x = np.array([35,25,23,15])

labels = ['apple','orange','banana','cherries']
c = ['red','orange','yellow' ,'pink']

plt.pie(x ,labels = labels,colors = c)
plt.legend()

plt.show()
```
