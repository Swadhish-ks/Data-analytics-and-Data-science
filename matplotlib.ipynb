```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])

plt.plot(ypoints ,marker = 'o')
plt.show()
```

```python
ypoints = np.array([3,8,1,10])

plt.plot(ypoints ,linestyle = 'dotted')
plt.show()
```

```python
ypoints = np.array([3,8,1,10])

plt.plot(ypoints ,color = 'r')
plt.show()
```

```python

x = np.array(['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'])
y = np.array([100,156,140,120,115,150,170,112,113,114,200,134])

plt.plot(x,y)

plt.title("Stock price data")
plt.xlabel("MONTHS(JAN - DEC)")
plt.ylabel("price")
plt.xticks(rotation = 60)
plt.grid()
plt.show()
```

#multiple plot

2 plot 

```python
#plot 1 :
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.grid()

#plot 2 :
x = np.array([0,1,2,3,])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.grid()

plt.show()
```

#4 plot

```python
#plot 1 :
x = np.array([0,1,2,3])
y = np.array([9,8,1,11])



plt.subplot(1,2,1)
plt.xlabel("item")
plt.ylabel("price")
plt.plot(x,y)
plt.title("sales")
plt.grid()




#plot 2 :
x = np.array([0,1,2,3,])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.xlabel("person")
plt.ylabel("income tax")
plt.plot(x,y)
plt.grid()
plt.title("income")



plt.show()
```
